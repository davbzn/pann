```python
def PeakPowerAtWavelength( data_x, data_y, mute = False):
    peakPower = max(data_y)
    indexPeakPower = data_y.tolist().index(peakPower)
    peakWavelength = data_x[indexPeakPower]
    if (not mute):
        print('peak', peakPower, 'mW at', peakWavelength, 'nm')
    return peakPower, peakWavelength

peak_wlen = []

print('Pump power ~1mW:')
pwr, wlen = PeakPowerAtWavelength(d['T1@52'][:,0],d['T1@52'][:,1])
peak_wlen.append( 1552-wlen )
pwr, wlen = PeakPowerAtWavelength(d['T1@54'][:,0],d['T1@54'][:,1])
peak_wlen.append( 1554-wlen )
pwr, wlen = PeakPowerAtWavelength(d['T1@56'][:,0],d['T1@56'][:,1])
peak_wlen.append( 1556-wlen )

 = np.mean(peak_wlen)
print('mean wavelength shift is\n =',,'nm')

del pwr, wlen
```
