```python
def PeakPowerAtWavelength( data_x, data_y, mute = False):
    peakPower = max(data_y)
    indexPeakPower = data_y.tolist().index(peakPower)
    peakWavelength = data_x[indexPeakPower]
    if (not mute):
        print('peak', peakPower, 'mW at', peakWavelength, 'nm')
    return peakPower, peakWavelength

peak_wlen = []

print('Pump power ~10mW:')
pwr, wlen = PeakPowerAtWavelength(d['M3P10_osa_1'][:,0],d['M3P10_osa_1'][:,1])
peak_wlen.append( wlen )
pwr, wlen = PeakPowerAtWavelength(d['M3P10_osa_2'][:,0],d['M3P10_osa_2'][:,1])
peak_wlen.append( wlen )
print()
print('Pump power ~20mW:')
pwr, wlen = PeakPowerAtWavelength(d['M3P20_osa_1'][:,0],d['M3P20_osa_1'][:,1])
peak_wlen.append( wlen )
pwr, wlen = PeakPowerAtWavelength(d['M3P20_osa_2'][:,0],d['M3P20_osa_2'][:,1])
peak_wlen.append( wlen )
print()
print('Pump power ~30mW:')
pwr, wlen = PeakPowerAtWavelength(d['M3P30_osa_1'][:,0],d['M3P30_osa_1'][:,1])
peak_wlen.append( wlen )
pwr, wlen = PeakPowerAtWavelength(d['M3P30_osa_2'][:,0],d['M3P30_osa_2'][:,1])
peak_wlen.append( wlen )
print()
print('Pump power ~40mW:')
pwr, wlen = PeakPowerAtWavelength(d['M3P40_osa_1'][:,0],d['M3P40_osa_1'][:,1])
peak_wlen.append( wlen )
pwr, wlen = PeakPowerAtWavelength(d['M3P40_osa_2'][:,0],d['M3P40_osa_2'][:,1])
peak_wlen.append( wlen )
print()
print('Pump power ~50mW:')
pwr, wlen = PeakPowerAtWavelength(d['M3P50_osa_1'][:,0],d['M3P50_osa_1'][:,1])
peak_wlen.append( wlen )
pwr, wlen = PeakPowerAtWavelength(d['M3P50_osa_2'][:,0],d['M3P50_osa_2'][:,1])
peak_wlen.append( wlen )
print()
print('Pump power ~60mW:')
pwr, wlen = PeakPowerAtWavelength(d['M3P60_osa_1'][:,0],d['M3P60_osa_1'][:,1])
peak_wlen.append( wlen )
pwr, wlen = PeakPowerAtWavelength(d['M3P60_osa_2'][:,0],d['M3P60_osa_2'][:,1])
peak_wlen.append( wlen )
print()
print('Pump power ~01mW:')
pwr, wlen = PeakPowerAtWavelength(d['M3P01_osa_1'][:,0],d['M3P01_osa_1'][:,1])
peak_wlen.append( wlen )
pwr, wlen = PeakPowerAtWavelength(d['M3P01_osa_2'][:,0],d['M3P01_osa_2'][:,1])
peak_wlen.append( wlen )
print()

 = 1552.5-np.mean(peak_wlen)
print('mean wavelength shift is\n = 1552.5 -', np.mean(peak_wlen),'nm =', ,'nm')
```
